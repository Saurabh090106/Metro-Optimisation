```python
path = kagglehub.dataset_download("shivamb/real-or-fake-fake-jobposting-prediction")
print("Path to dataset files:", path)
import numpy as np
import pandas as pd
```

```python
dataset_path = path
csv_file =  os.path.join(dataset_path, "fake_job_postings.csv")
df = pd.read_csv(csv_file)
```

```python
df.head()
```

```python
print(df.isna().sum())
df.shape
```

```python
df = df.dropna()
df.shape
```

```python
column_to_clean = df[["company_profile","description","requirements","benefits"]]
```

```python
def remove_stopwords(text):
  st_words = set(stopwords.words("english"))
  word_tokens = word_tokenize(text)
  filtered_text = [word for word in word_tokens if word.lower() not in st_words]
  return filtered_text
for col in column_to_clean:
  df[col]=df[col].apply(remove_stopwords)
```

```python
import string
def remove_punctuation(text):
    result = ''.join([char.lower() for char in text if char not in string.punctuation])
    return result
for col in column_to_clean:
  df[col]=df[col].apply(remove_punctuation)
```
